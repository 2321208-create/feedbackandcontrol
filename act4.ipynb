```python
import control as ctrl
import matplotlib.pyplot as plt

# Prob 1
num1 = [1]
den1 = [1, 2, 2]

G = ctrl.tf(num1, den1)
print(G)

poles1 = ctrl.poles(G)
print("No. 1 Poles:", poles1)

zeroes1 = ctrl.zeros(G)
print("No. 1 Zeroes:", zeroes1)

ctrl.pzmap(G, grid=True)
plt.show()

# Prob 2
num2 = [1]
den2 = [1, 2, 1]

G = ctrl.tf(num2, den2)
print(G)

poles2 = ctrl.poles(G)
print("No. 2 Poles:", poles2)

zeroes2 = ctrl.zeros(G)
print("No. 2 Zeroes:", zeroes2)

ctrl.pzmap(G, grid=True)
plt.show()
```
